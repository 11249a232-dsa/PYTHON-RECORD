```python
Aim = 'Prints whether the number is a "Palindrome" or "Not Palindrome"'
print(Aim)
number = input("Enter a number: ")
if number == number[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
for i in range(10):
    count = number.count(str(i))
    if count > 0:
        print(f"{i} appears {count} times")
```
